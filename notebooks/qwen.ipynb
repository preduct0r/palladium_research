```python
import os
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

dotenv_path = find_dotenv()
if not dotenv_path:
    raise FileNotFoundError("'.env' file not found. Please create one in the project root.")
load_dotenv(dotenv_path)

token = os.getenv("YANDEX_API_KEY")
folder_id = os.getenv("YANDEX_FOLDER_ID")
```

```python
client = OpenAI(
        api_key=token,
        base_url="https://llm.api.cloud.yandex.net/v1"
    )

completion = client.chat.completions.create(
    model=f"gpt://{folder_id}/qwen3-235b-a22b-fp8/latest",
    messages=[
        {"role": "developer", "content": "–¢—ã –æ—á–µ–Ω—å —É–º–Ω—ã–π –∞—Å—Å–∏—Å—Ç–µ–Ω—Ç."},
        {
            "role": "user",
            "content": "–ß—Ç–æ —É–º–µ—é—Ç –±–æ–ª—å—à–∏–µ —è–∑—ã–∫–æ–≤—ã–µ –º–æ–¥–µ–ª–∏?",
        },
    ],
)

print(completion.choices[0].message.content)
```
